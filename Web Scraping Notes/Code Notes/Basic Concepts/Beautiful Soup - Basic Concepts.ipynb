```python
from bs4 import BeautifulSoup
import requests
```

## HTTP Request

#### make a request

```python
response = requests.get('http://automationpractice.com/index.php') 
```

#### check status code

```python
response.status_code
```

## Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

## Important functions

#### find()

```python
contact_us = soup.find(id = 'contact-link').find('a').get_text()
contact_us
```

```python
contact_us = soup.find(id = 'contact-link').get_text().replace('\n', '')
contact_us
```

```python
popular = soup.find(class_ = 'homefeatured').get_text()
popular
```

```python
first_element = soup.find(class_ = 'ajax_block_product').find(class_ = 'content_price')
first_element
```

#### find_all() & findAll() => returns a list

```python
result = soup.findAll(class_ = 'ajax_block_product')
result
```

```python
len(result)
```

#### select_one() => equivalent to find()

```python
soup.select_one('.shop-phone').get_text().strip().replace('Call us now: ', '')
```

```python
soup.select_one('#contact-link').get_text().strip()
```

#### select() => equivalent to findAll()

```python
result_s = soup.select('.ajax_block_product')
result_s
```

```python
len(result_s)
```

#### get value of attributes vs get_text()

```python
soup.find(itemprop="name").find('a').get('title')
```

```python
soup.find(itemprop="name").find('a').get_text().strip()
```

## Navigate to Siblings & Parents

```python
soup.find(class_ = 'ajax_block_product').find_next_sibling().find(itemprop = "price").get_text().strip()
```

```python
soup.findAll(class_ = 'ajax_block_product')[-1].find(itemprop = "price").get_text().strip()
```

```python
soup.findAll(class_ = 'ajax_block_product')[-1].find_previous_sibling().find(itemprop = "price").get_text().strip()
```

```python
soup.findAll(class_ = 'ajax_block_product')[1].find(itemprop = "price").get_text().strip()
```

```python
soup.find(class_ = 'ajax_block_product').find_parent()
```

## Extract Links and URLs

### extract single link

```python
soup.find(id = "contact-link").find('a')['href']
```

### extract multiple links

```python
ul = soup.find('ul')
```

```python
for a in ul.find_all('a'):
  print(a['href'])
```
