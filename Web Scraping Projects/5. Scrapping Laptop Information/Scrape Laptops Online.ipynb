### Pagenation

```python
product_name = []
rating = []
CPU = []
video_card = []
RAM = []
SSD = []
display = []
price = []
string = []

for i in range(1,20):
    website = 'https://ardes.bg/laptopi/laptopi/page/' + str(i)
    response = requests.get(website)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.findAll('div', {'class':"product"}) 

    for result in results:
        try:
            product_name.append(result.find('div', {'class':"isTruncated"}).get_text())
        except:
            product_name.append('n/a')

        try:
            rating.append(result.find('div', {'class': "rating col-md-9"}).span.get_text())
        except:
            rating.append('n/a')

        try:
            CPU.append(result.findAll('li')[0].get_text())
        except:
            CPU.append('n/a')

        try:
            video_card.append(result.findAll('li')[1].get_text())
        except:
            video_card.append('n/a')

        try:
            RAM.append(result.findAll('li')[2].get_text())
        except:
            RAM.append('n/a')

        try:
            SSD.append(result.findAll('li')[3].get_text()) 
        except:
            SSD.append('n/a') 

        try:
            display.append(result.findAll('li')[4].get_text())
        except:
            display.append('n/a')

        try:
             price.append(result.find('span', {'class':"price-num"}).get_text())
        except:
            price.append('n/a')

        try:    
            string.append(result.find('a')['href'])
        except:
            string.append('n/a')
```

```python
link = []
for text in string:  
    if text != 'n/a':    
        link.append(urljoin(root, text))
    else:
        link.append('n/a')
```

```python
laptops_df = pd.DataFrame({'Product': product_name, 'Rating': rating, 'CPU': CPU, 'Video_card': video_card,
             'RAM': RAM, 'SSD': SSD, 'Display': display, 'Price': price, 'Link': link})
```

```python
laptops_df.head()
```

### Data Cleaning

```python
laptops_df['Rating'] = laptops_df['Rating'].apply(lambda x: x.strip('(').strip(')'))
laptops_df.head()
```

```python
prices = pd.Series(laptops_df['Price'])
new_prices = []
for price in prices:
    base = price[:-4]
    length = len(price)
    last_two_chars = price[length-4:]
    concat_string = base + ',' + last_two_chars
    new_prices.append(concat_string)
new_prices
```

```python
laptops_df['Price'] = new_prices
laptops_df.head()
```

```python
laptops_df.to_excel('laptops.xlsx', index = False)
```
